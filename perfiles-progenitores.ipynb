```python
def sersic_charly_ie(r,ie,re,beta):
    return ie*np.exp(-(2./beta-0.33)*((r/re)**beta-1))
```

```python
def sersic_charly_mue(r,mue,re,beta):
    return mue + 2.5*(2./beta-0.33)*((r/re)**beta -1)/np.log(10)
```

```python
def mu(_ie):
    return -2.5*np.log10(_ie)-48.6 -0.044
```
