```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText

m026cp = np.loadtxt('../dust/026_dust_band125.txt'  ,usecols=[4], unpack=True)
m026sp = np.loadtxt('026_nodust_band125.txt',usecols=[4], unpack=True)
m032   = np.loadtxt('032_nodust_band125.txt',usecols=[4], unpack=True)
m041   = np.loadtxt('041_nodust_band125.txt',usecols=[4], unpack=True)
m091   = np.loadtxt('091_nodust_band125.txt',usecols=[4], unpack=True)
        
```
